```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dados = pd.read_csv('evdataset.csv')
```

```python
dados.isnull().sum()
```

```python
ax,fig = plt.subplots(figsize =(10,10))
ax = sns.countplot(dados['Seats'])
```

```python
dados_comparativos = dados[['Acceleration 0 - 100 km/h', 'Battery Capacity', 'Seats', 'Height', 'Max. Payload']]
```

```python
sns.pairplot(dados_comparativos, hue = 'Seats')
```

```python
sns.heatmap(dados_comparativos.corr())
```
